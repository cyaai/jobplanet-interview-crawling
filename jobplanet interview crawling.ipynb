```python
from selenium import webdriver
import time
import datetime
import pandas as pd

browser = webdriver.Firefox()
browser.get('https://www.jobplanet.co.kr/')
browser.maximize_window()
browser.implicitly_wait(6)  # seconds
```

```python
# Click the link to go to the login form
login_link = browser.find_element_by_css_selector('a.btn_txt.login')
login_link.click()
```

```python
# id
username_input = browser.find_element_by_css_selector('input#user_email')
username_input.send_keys("id") 
```

```python
# password
password_input = browser.find_element_by_css_selector('input#user_password')
password_input.send_keys("password") 

browser.find_element_by_css_selector('input.btn_sign_up').click()
```

```python
browser.get('https://www.jobplanet.co.kr/companies/1186/interviews?page=1') # 현대해상
```

```python
jobplanet_df = pd.DataFrame(columns=["role", "date", "level", "main", "question", "answer"], )
```

```python
for board_number in range(1,20):
    browser.get("https://www.jobplanet.co.kr/companies/1186/interviews?page=&page=" + str(board_number))
    time.sleep(0.1)
    
    contents = browser.find_elements_by_css_selector('section.content_ty4.content_ty4_1.video_ad_content')

    for i in range(len(contents)):
        try:
            role = contents[i].find_element_by_css_selector("span.txt1").text.split("|")[0].split("  / ")[0]
            date = contents[i].find_element_by_css_selector("span.txt1").text.split("| ")[-1]
            level = contents[i].find_element_by_css_selector("span.blo_txt2.us_txt_mr").text # 면접 난이도
            main = contents[i].find_element_by_css_selector("h2.us_label ").text
            question = contents[i].find_element_by_css_selector("dd.df1").text # 면접 질문
            answer = contents[i].find_element_by_css_selector("span.answer.mobile_full_content.notranslate").text # 면접 질문 
            
        except:
            answer = ""

        jobplanet_df.loc[len(jobplanet_df)] = [role, date, level, main, question, answer]

    time.sleep(1)
```
